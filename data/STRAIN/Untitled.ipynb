```python
import pandas as pd
import matplotlib.pyplot as plt

# Load train and test CSV files
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
labels = pd.read_csv('labels.csv')

# Print the contents of train dataframe
print("Train Data:")
print(train_df.head())  # Display the first few rows of the train dataframe
print()

# Print the contents of test dataframe
print("Test Data:")
print(test_df.head())  # Display the first few rows of the test dataframe
print()

# Print the contents of labels dataframe
print("Labels Data:")
print(labels.head())  # Display the first few rows of the labels dataframe
print()

# Plot the first five test samples
for i in range(5):
    plt.figure()
    plt.plot(test_df.iloc[i])  # Plot the time series data for the i-th test sample
    plt.title(f"Test Sample {i+1} - Label: {labels.iloc[i, 0]}")  # Assuming labels are in the first column
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.show()
```
